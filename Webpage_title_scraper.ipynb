```python
from bs4 import BeautifulSoup 
import re 
import urllib.request as urllib2
import urllib 
import urllib.error
import pandas as pd 
import numpy as np
import os
os.environ['http_proxy']=''
# read urls of websites from text file 
browser_history = pd.read_csv('/content/URLs - bq-results-20220421-181700-1650565178544.csv')
browser_history = browser_history.dropna()

urls = browser_history['value'].to_list()
browser_history.head()


```

```python
import requests

try:
    response = requests.get('https://www.indigocard.com/platinum-application/SSA-ITA?CID=400000&mmcid=IAAG00717052001?22875739525#', timeout=5, verify= False)
    response.raise_for_status()
    # Code here will only run if the request is successful
except requests.exceptions.HTTPError as errh:
    print(errh)
except requests.exceptions.ConnectionError as errc:
    print(errc)
except requests.exceptions.Timeout as errt:
    print(errt)
except requests.exceptions.RequestException as err:
    print(err)
except:
    pass    

soup = BeautifulSoup(response.content, "html5lib")
if soup.title != None:
    print(soup.find_all('|<h[^>]+>(.*)</h[^>]+>|iU'))
else: 
    print('no title')    
```

```python
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
c_headers={'User-Agent':user_agent} 
parsed_text = []
for url in urls: 
    try:

        response = requests.get(url, timeout=5, headers = c_headers)
        response.raise_for_status()
        
        # Code here will only run if the request is successful
    except requests.exceptions.HTTPError as errh:
        print(errh)
    except requests.exceptions.ConnectionError as errc:
        print(errc)
    except requests.exceptions.Timeout as errt:
        print(errt)
    except requests.exceptions.RequestException as err:
        print(err)
    soup = BeautifulSoup(response.content, "html5lib")
    if soup.title != None:
        parsed_text.append(soup.title.string)
    else: 
        parsed_text.append('no title')
    
        
    
        
```

```python
output = pd.DataFrame(parsed_text)
```
